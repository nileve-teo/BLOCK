En la ingeniería estructural de rascacielos frente a sismos, se utilizan disipadores viscosos colocados en las uniones de las vigas. Cuando el edificio se mueve, el fluido dentro del amortiguador genera una fuerza de oposición.

Para movimientos severos, la resistencia del fluido deja de ser lineal y se modela mediante la ecuación de fricción cuadrática de Van der Pol:

<p align="center">

[
\frac{d^2x}{dt^2}-\mu(1-x^2)\frac{dx}{dt}+x=0
]

</p>

Donde (x(t)) representa el desplazamiento normalizado de la estructura y (\mu = 1.5) es el coeficiente de amortiguamiento no lineal.

Suponga que una réplica sísmica inicial desplaza la estructura de su eje central, dejando las condiciones iniciales:

<p align="center">

[
x(0)=2
]

[
\dot{x}(0)=0
]

</p>
