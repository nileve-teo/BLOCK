<p>
En la ingeniería estructural de rascacielos frente a sismos, se utilizan disipadores viscosos colocados en las uniones de las vigas. Cuando el edificio se mueve, el fluido dentro del amortiguador genera una fuerza de oposición.
</p>

<p>
Para movimientos severos, la resistencia del fluido deja de ser lineal y se modela mediante la ecuación de fricción cuadrática de Van der Pol:
</p>

<div align="center">

$$
\frac{d^2x}{dt^2}-\mu(1-x^2)\frac{dx}{dt}+x=0
$$

</div>

<p>
Donde \(x(t)\) representa el desplazamiento normalizado de la estructura y \(\mu = 1.5\) es el coeficiente de amortiguamiento no lineal.
</p>

<p>
Suponga que una réplica sísmica inicial desplaza la estructura de su eje central, dejando las condiciones iniciales:
</p>

<div align="center">

$$
x(0)=2   
$$

$$
\dot{x}(0)=0   
$$

</div>
<p>
Siendo 2 el desplazamiento   máximo y 0 el reposo instantáneo.
</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

# =====================================================
# PARÁMETROS
# =====================================================

mu = 1.5
t0 = 0
tf = 20
h = 0.05

# =====================================================
# FUNCIÓN VAN DER POL
# =====================================================

def f(x, v):
    dxdt = v
    dvdt = mu*(1 - x**2)*v - x
    return dxdt, dvdt

# =====================================================
# FUNCIÓN RK4
# =====================================================

def RK4(h):

    N = int((tf-t0)/h)

    t = np.linspace(t0, tf, N+1)

    x = np.zeros(N+1)
    v = np.zeros(N+1)

    x[0] = 2.0
    v[0] = 0.0

    for i in range(N):

        k1x, k1v = f(x[i],v[i])

        k2x, k2v = f(
            x[i] + h*k1x/2,
            v[i] + h*k1v/2
        )

        k3x, k3v = f(
            x[i] + h*k2x/2,
            v[i] + h*k2v/2
        )

        k4x, k4v = f(
            x[i] + h*k3x,
            v[i] + h*k3v
        )

        x[i+1] = x[i] + h*(k1x + 2*k2x + 2*k3x + k4x)/6
        v[i+1] = v[i] + h*(k1v + 2*k2v + 2*k3v + k4v)/6

    return t,x,v

# =====================================================
# SOLUCIÓN PRINCIPAL
# =====================================================

t,x,v = RK4(0.05)

# aceleración
a = mu*(1-x**2)*v - x

# =====================================================
# TABLA PRIMERAS 10 ITERACIONES
# =====================================================

tabla10 = pd.DataFrame({
    "Iter": np.arange(10),
    "t [s]": np.round(t[:10],4),
    "x(t)": np.round(x[:10],6),
    "v(t)": np.round(v[:10],6)
})

print("\n"+"="*70)
print("PRIMERAS 10 ITERACIONES")
print("="*70)
print(tabla10.to_string(index=False))

# =====================================================
# TABLA ESTADOS FINALES
# =====================================================

tabla_final = pd.DataFrame({
    "Variable":["Tiempo final","Desplazamiento final","Velocidad final"],
    "Valor":[
        round(t[-1],4),
        round(x[-1],6),
        round(v[-1],6)
    ]
})

print("\n"+"="*70)
print("ESTADO FINAL")
print("="*70)
print(tabla_final.to_string(index=False))

# =====================================================
# FIGURA 1
# EVOLUCIÓN TEMPORAL
# =====================================================

fig,ax1 = plt.subplots(figsize=(12,6))

ax1.plot(
    t,
    x,
    linewidth=2.5,
    label='Desplazamiento x(t)'
)

ax1.plot(
    t,
    v,
    linewidth=2.5,
    label='Velocidad v(t)'
)

# Región de aceleración alta
idx = np.argmax(np.abs(a))

ax1.axvspan(
    max(0,t[idx]-0.5),
    min(tf,t[idx]+0.5),
    alpha=0.25
)

ax1.scatter(
    t[idx],
    x[idx],
    s=100
)

ax1.annotate(
    f'Máxima |a| = {abs(a[idx]):.2f}',
    xy=(t[idx],x[idx]),
    xytext=(t[idx]+1,x[idx]+1),
    arrowprops=dict(arrowstyle='->')
)

ax1.set_title(
    'Evolución Temporal del Oscilador de Van der Pol',
    fontsize=14,
    fontweight='bold'
)

ax1.set_xlabel('Tiempo [s]')
ax1.set_ylabel('Magnitud')

ax1.grid(True,which='both',alpha=0.3)

ax1.minorticks_on()

ax1.legend()

plt.tight_layout()
plt.show()

# =====================================================
# FIGURA 2
# ESPACIO DE FASES MEJORADO
# =====================================================

fig,ax = plt.subplots(figsize=(8,8))

points = np.array([x,v]).T.reshape(-1,1,2)
segments = np.concatenate(
    [points[:-1],points[1:]],
    axis=1
)

lc = LineCollection(
    segments,
    cmap='viridis',
    linewidth=2
)

lc.set_array(t)

ax.add_collection(lc)

ax.scatter(
    x[0],
    v[0],
    s=120,
    marker='o',
    label='Inicio (2,0)'
)

ax.scatter(
    x[-1],
    v[-1],
    s=120,
    marker='s',
    label='Final'
)

# Flechas de dirección temporal
saltos = 40

for i in range(0,len(x)-saltos,saltos):

    ax.arrow(
        x[i],
        v[i],
        x[i+1]-x[i],
        v[i+1]-v[i],
        shape='full',
        head_width=0.08,
        length_includes_head=True,
        alpha=0.5
    )

ax.set_xlim(min(x)-0.5,max(x)+0.5)
ax.set_ylim(min(v)-0.5,max(v)+0.5)

ax.set_xlabel('Desplazamiento x')
ax.set_ylabel('Velocidad v')

ax.set_title(
    'Diagrama de Espacio de Fases',
    fontsize=14,
    fontweight='bold'
)

ax.grid(True)

ax.legend()

cbar = plt.colorbar(lc)
cbar.set_label('Tiempo [s]')

plt.tight_layout()
plt.show()

# =====================================================
# FIGURA 3
# PRUEBA DE ESTABILIDAD NUMÉRICA
# =====================================================

t1,x1,v1 = RK4(0.05)
t2,x2,v2 = RK4(0.5)
t3,x3,v3 = RK4(1.0)

plt.figure(figsize=(12,6))

plt.subplot(311)
plt.plot(t1,x1)
plt.title("h = 0.05 s")

plt.subplot(312)
plt.plot(t2,x2)
plt.title("h = 0.5 s")

plt.subplot(313)
plt.plot(t3,x3)
plt.title("h = 1.0 s")

plt.yscale("symlog")

plt.tight_layout()
plt.show()

# =====================================================
# TABLA DE DIVERGENCIA
# =====================================================

print("\n")
print("="*70)
print("PRUEBA DE ESTABILIDAD NUMÉRICA")
print("="*70)

print(f"h = 0.05  --> x_final = {x1[-1]:.6f}")
print(f"h = 0.50  --> x_final = {x2[-1]:.6f}")
print(f"h = 1.00  --> x_final = {x3[-1]:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 1.5
t0 = 0
tf = 20
h = 0.05

# Número de pasos
N = int((tf - t0)/h)

# Vector tiempo
t = np.linspace(t0, tf, N+1)

# Sistema de Van der Pol
def f(x, v):
    dxdt = v
    dvdt = mu*(1 - x**2)*v - x
    return dxdt, dvdt

# Condiciones iniciales
x = np.zeros(N+1)
v = np.zeros(N+1)

x[0] = 2.0
v[0] = 0.0

# Método RK4
for i in range(N):

    k1x, k1v = f(x[i], v[i])

    k2x, k2v = f(
        x[i] + h*k1x/2,
        v[i] + h*k1v/2
    )

    k3x, k3v = f(
        x[i] + h*k2x/2,
        v[i] + h*k2v/2
    )

    k4x, k4v = f(
        x[i] + h*k3x,
        v[i] + h*k3v
    )

    x[i+1] = x[i] + h*(k1x + 2*k2x + 2*k3x + k4x)/6
    v[i+1] = v[i] + h*(k1v + 2*k2v + 2*k3v + k4v)/6

# Tabla de datos
print("Primeras 10 iteraciones")
print("Iter\tTiempo\t\tx\t\tv")

for i in range(10):
    print(f"{i}\t{t[i]:.2f}\t\t{x[i]:.6f}\t{v[i]:.6f}")

print("\nEstado final")
print(f"t = {t[-1]:.2f} s")
print(f"x = {x[-1]:.6f}")
print(f"v = {v[-1]:.6f}")

# Gráfica 1
plt.figure(figsize=(10,5))
plt.plot(t,x,label='x(t)',linewidth=2)
plt.plot(t,v,label='v(t)',linewidth=2)
plt.xlabel('Tiempo (s)')
plt.ylabel('Magnitud')
plt.title('Evolución Temporal')
plt.grid(True)
plt.legend()
plt.show()

# Gráfica 2
plt.figure(figsize=(6,6))
plt.plot(x,v,'b')
plt.xlabel('x')
plt.ylabel('v')
plt.title('Diagrama de Espacio de Fases')
plt.grid(True)
plt.show()
```
